## Use any 2 types of charts to visualize some data

#### you can use datasets or you can write your own data

```python
data = {'year': [2010, 2011, 2012, 2010, 2011, 2012, 2010, 2011, 2012],
        'team': ['FCBarcelona', 'FCBarcelona', 'FCBarcelona', 'RMadrid', 'RMadrid', 'RMadrid', 'ValenciaCF',
                 'ValenciaCF', 'ValenciaCF'],
        'wins':   [30, 28, 32, 29, 32, 26, 21, 17, 19],
        'draws':  [6, 7, 4, 5, 4, 7, 8, 10, 8],
        'losses': [2, 3, 2, 4, 2, 5, 9, 11, 11]}
```

```python
df = pd.DataFrame(data, columns = ['year', 'team', 'wins', 'draws', 'losses'])
```

```python
tem = ['FCBarcelona', 'RMadrid', 'ValenciaCF']
win = [32, 26, 19]

plt.bar(tem, win, color = ['r', 'k', 'y'])

plt.title('Top 3 wins in 2012')
plt.xlabel('Team')
plt.ylabel('Wins');
```

```python
plt.scatter(df['team'], df['losses'], color = 'red');

plt.xlabel('Team')
plt.ylabel('Losses');
```

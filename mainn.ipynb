```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv('tmdb-movies.csv')
df
```

```python
df.info()
```

```python
df = df.drop(['cast','homepage', 'director', 'tagline', 'keywords','overview',
              'production_companies','release_date','budget_adj','revenue_adj'], axis = 1)
```

```python
df.head()
```

```python
plt.hist(df['vote_average']);
```

```python
plt.hist(df['vote_average'], facecolor = 'b', edgecolor = 'r', bins = 10)
plt.title('Distribution of votes', fontsize = 15)
plt.ylabel('Movies count')
plt.xlabel('Votes');
```

```python
film_yearly_cnt = df['release_year'].value_counts(ascending = True)
film_yearly_cnt
```

```python
years = df.groupby('release_year', as_index = False).sum()['release_year']
years
```

```python
plt.plot(years, film_yearly_cnt);
```

```python
plt.plot(years, film_yearly_cnt, c = 'black')
plt.title('Movies produced along years', fontsize = 15)
plt.ylabel('Movies count')
plt.xlabel('years')

plt.xlim(1960, 2010)
plt.xticks(range(1960, 2011, 5));
```
